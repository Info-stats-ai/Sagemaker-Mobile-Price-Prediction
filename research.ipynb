### Script used by AWS Sagemaker To train Models

```python
%%writefile script.py 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score  
import sklearn
import joblib
import boto3
import pathlib
from io import StringIO
import argparse
import os
import sys
import pandas as pd
import numpy as np


def model_fn(model_dir):
    model = joblib.load(os.path.join(model_dir, "model.joblib"))

    if __name__ == "__main__":
        print("....[INFO] model_fn.....")
        # parse arguments
        # what it does is it parses the arguments from the command line
        # and it is used to pass the arguments to the scripts 
        #Hyperparameters thesea are required by sagemaker
        parser = argparse.ArgumentParser()
        parser.add_argument("--n_estimators", type=int, default=100)
        parser.add_argument("--random_state", type=int, default=42)
        args = parser.parse_args()
        
        # Data, model and output directories 
        parser.add_argument("--train", type=str, default=os.environ["SM_CHANNEL_TRAIN"])
        
        parser.add_argument("--test", type=str, default=os.environ["SM_CHANNEL_TEST"])
        # The directory where the model will be saved
        parser.add_argument("--model_dir", type=str, default=os.environ["SM_MODEL_DIR"])
       
        parser.add_argument("--output_dir", type=str, default=os.environ["SM_OUTPUT_DIR"])
        parser.add_argument("--train-file", type=str, default= "train-V-1.csv")
        parser.add_argument("--test-file", type=str, default= "test-V-1.csv")

        args, _ = parser.parse_known_args()
        print(f"Received arguments {args}")
        print("SKLEARN VERSION: ", sklearn.__version__)
        print("JOBLIB VERSION: ", joblib.__version__)
        print("BOTO3 VERSION: ", boto3.__version__)
        print("PATHLIB VERSION: ", pathlib.__version__)
        print("ARGPARSE VERSION: ", argparse.__version__)
        print("OS VERSION: ", os.__version__)
        print("SYSY VERSION: ", sys.__version__)
        

        # Read data
        train_data = pd.read_csv(os.path.join(args.train, args.train_file))
        test_data = pd.read_csv(os.path.join(args.test, args.test_file))

        # Feature names
        feature_names = train_data.columns.drop("price_range")
        label_name = "price_range"
        print("....Buidling training and testing data...")
        X_train = train_data[feature_names]
        y_train = train_data[label_name]
        X_test = test_data[feature_names]
        y_test = test_data[label_name]

        print("....Training model.....")
        # Define and train model
        model = RandomForestClassifier(
            n_estimators=args.n_estimators,
            random_state=args.random_state, verbose=2,
            n_jobs=1
        )
        # fit model
        model.fit(X_train, y_train)

        # Save model
        joblib.dump(model, os.path.join(args.model_dir, "model.joblib"))

        joblib.dump(model, model_path)
        print("....Model saved at {}".format(model_path))
```
